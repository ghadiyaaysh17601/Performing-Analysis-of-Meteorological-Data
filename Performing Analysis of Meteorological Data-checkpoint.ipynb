# Performing Analysis of Meteorological Data

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('weatherHistory.csv',parse_dates=['Formatted Date'])
df.head()
```

```python
df.info
```

```python
df.shape
```

```python
df=df.drop(['Summary','Precip Type','Temperature (C)','Wind Speed (km/h)','Wind Bearing (degrees)','Visibility (km)','Loud Cover','Pressure (millibars)','Daily Summary'],axis=1)
```

```python
df.head()
```

```python
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], errors='coerce',utc=True)

df.head()
```

```python
df
```

```python
df.set_index('Formatted Date').head()
```

```python
df2 = df.resample('M', on='Formatted Date').mean()
df2
```

# April

```python
plt.figure(figsize=(15,7))
plt.style.use('fivethirtyeight')
plt.plot(df2['Humidity'],label='Humidity',color='red')
plt.plot(df2['Apparent Temperature (C)'],label='Apparent Temperature (C)',color='green')
plt.title('Variation of Apparent Temperature VS Humidity in 10 years',fontsize=23)
plt.legend(fontsize=15)
plt.show()
```

```python
april=df2[df2.index.month==4]
april
```

```python
plt.figure(figsize=(15,7))
plt.style.use('fivethirtyeight')
plt.plot(april.loc['2006-04-01':'2016-04-01','Apparent Temperature (C)'],label='Apparent Temperature (C)',marker='o')
plt.plot(april.loc['2006-04-01':'2016-04-01','Humidity'],label='Humidity',marker='o')
plt.title('Temperature VS Humidity',fontsize=23)
plt.xlabel('Month of April',fontsize=15)
plt.legend(fontsize=15)
plt.show()

```
